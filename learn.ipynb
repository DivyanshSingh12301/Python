# Handling Null values

```python
coffee.head()
```

```python
# If we want to show any block to be NAN(NaN stands for Not a Number and is the primary sentinel value used to represent missing or unavailable data in numeric columns.)
coffee.loc[[0,1], "Units Sold"] = np.nan
```

```python
coffee.head()
```

```python
# If we want to see number of explicit NAN values then for that
coffee.isna().sum()
```

```python
# Now if there is a dataset and some specific blocks have NAN value and we want to fill it then
coffee = coffee.fillna(1000)
```

```python
coffee.head()
```

```python
# But it is not that logical and if the dataset is big and we can not find proper value to fill then we can use
coffee.fillna(coffee['Units Sold'].mean()) # It is a much more realistic value
```

```python
# And if there is a pattern to the data then we can contine filling by that pattern by 
coffee.fillna(coffee['Units Sold'].interpolate())
# But it will not work because Interpolate only starts evaluating pattern from the beginning anf if the index 0 or 1 are emtpy then it cant predict the sequence
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate().bfill()
# Or coffee.fillna(coffee['Units Sold'].interpolate().bfill()) Bfill means backward fill
```

```python
coffee
```

```python
# If we want to drop the whole row which has NAN then coffee.dropna()
coffee.dropna().head()
```

```python
# If we want to drop value in which some specific column has NAN value 
coffee.dropna(subset = ['Units Sold']).head()
```

```python
coffee.head()
```

```python
coffee[coffee['Units Sold'].isna()] # To see which columns and row have NAN values
```

```python
coffee[coffee['Units Sold'].notna()].head() # To see which columns and row do not have NAN values
```
